```python
import os
from PIL import Image
from pathlib import Path
```

```python
def resize_images(directory_path, target_size=(224, 224)):
    # Creating a Path object
    directory = Path(directory_path)
    
    # Getting all jpg files in the directory
    image_files = list(directory.glob('*.jpg'))
    
    for image_path in image_files:
        try:
            with Image.open(image_path) as img:
                resized_img = img.resize(target_size, Image.Resampling.LANCZOS)
                # Saving the resized image, overwriting the original
                resized_img.save(image_path, quality=95)
            print(f"Successfully resized {image_path}")
        except Exception as e:
            print(f"Error processing {image_path}: {str(e)}")

directories = [
    'test/images',
    'train/images',
    'valid/images'
]
for dir_path in directories:
    print(f"\nProcessing directory: {dir_path}")
    resize_images(dir_path)

print("\nImage resizing completed!")
```
